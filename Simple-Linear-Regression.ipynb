```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
import warnings
warnings.filterwarnings("ignore") 
```

```python
# Read the dataset
df = pd.read_csv("datasets/height-weight.csv")
df.head()
```

```python
# Scatterplot the data
sns.scatterplot(x=df['Weight'],y=df['Height'])
plt.title("Weight-Height")
plt.show()
```

```python
# Divide the dataset into independent features and target feature
X=df[['Weight']]
y=df[['Height']]
```

```python
# Make train and test data from dataset
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.20,random_state=42)
```

```python
# standardize the data
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
```

```python
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
plt.scatter(X_train,y_train)
```

```python
# Train the model
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
```

```python
regressor.fit(X_train,y_train)
```

```python
print("Coeficient",regressor.coef_)
print("Slope",regressor.intercept_)
```

```python
y_pred = regressor.predict(X_test)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
```

```python
mea = mean_absolute_error(y_test,y_pred)
mse = mean_squared_error(y_test,y_pred)
print("MEA", mea)
print("MSE", mse)
```

```python
from sklearn.metrics import r2_score
```

```python
r2 = r2_score(y_test,y_pred)
print(r2)
```

```python
adjusted_r2 = 1-(1-r2)*(len(y_test)-1)/(len(y_test)-X_test.shape[1]-1)
```
